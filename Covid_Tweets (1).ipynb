```python
df = df[['user_name', 'date', 'text']]
```

```python
def clean_text(text):
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Remove non-alphanumeric characters and symbols
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text
```

```python
print(df['text'].dtype)
print(df.info())
```

```python
df['text'] = df['text'].apply(clean_text)
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
df['tokens'] = df['text'].apply(word_tokenize)
```

```python
df['tokens'] = df['tokens'].apply(lambda x: [word.lower() for word in x])
```

```python
stop_words = set(stopwords.words('english'))
df['tokens'] = df['tokens'].apply(lambda x: [word for word in x if word not in stop_words])
```

```python
lemmatizer = WordNetLemmatizer()
df['tokens'] = df['tokens'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
import torch

model_dict = {
    "t5-small": (T5Tokenizer.from_pretrained("t5-small"), T5ForConditionalGeneration.from_pretrained("t5-small")),
    "t5-base": (T5Tokenizer.from_pretrained("t5-base"), T5ForConditionalGeneration.from_pretrained("t5-base")),
    "t5-large": (T5Tokenizer.from_pretrained("t5-large"), T5ForConditionalGeneration.from_pretrained("t5-large"))
}

# Choose a model from the dictionary or replace with your desired model name
chosen_model_name = "t5-large"

def summarize_tweet(tweet, model_name=chosen_model_name):

  tokenizer, model = model_dict.get(model_name)
  if tokenizer is None or model is None:
    tokenizer = T5Tokenizer.from_pretrained(model_name)
    model = T5ForConditionalGeneration.from_pretrained(model_name)
    model_dict[model_name] = (tokenizer, model)

  # Preprocess and tokenize the tweet
  inputs = tokenizer.encode("summarize: " + tweet, return_tensors="pt", max_length=1000, truncation=True)

  # Generate summary using the chosen T5 model
  summary_ids = model.generate(inputs, max_length=1000, num_beams=2, early_stopping=True)
  summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

  return summary
for i in range(0,5):
  tweet = input("Enter a tweet to summarize: ")
  summary = summarize_tweet(tweet)

  print("Original tweet:", tweet)
  print("Summary:", summary)

```
